```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Per salvare i dati usiamo le [Series di Pandas](http://pandas.pydata.org/pandas-docs/stable/api.html#series). 

```python
filename = '001_ServoCurveData_0000.nta'
```

```python
f = open(filename,'r')
lines = f.readlines()
```

```python
data = []
for line in lines:
    if not line.startswith('%'):
        data.append(line.replace('\n','').split('\t',8))
```

```python
v = np.ndarray(shape=(len(data),4),dtype=float)
for i in range(len(data)):
    for j in range(3,7):
        v[i-3,j-3]=float(data[i][j])
```

```python
stato = v[:,0] # Stato 0 inspirio - Stato 1 espirio
pressione = v[:,1] # Pressione
flusso = v[:,2] # Flusso
eadi = v[:,3] # EaDI
```

```python
index = np.nonzero(np.diff(stato)==1)[0]+1
```

```python
data_pressione = np.split(pressione,index)
```

```python
data_pressione[0:2]
```

Salviamo in un formato plain come il csv

```python
import csv

csvfile = "csvfile.csv"

with open(csvfile, "w") as output:
    writer = csv.writer(output, lineterminator='\n')
    for line in data_pressione:
        writer.writerow(line)    
```
